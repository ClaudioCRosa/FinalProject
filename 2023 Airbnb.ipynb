```python
#Loading up the relevant libraries and the data from Airbnb

import json
import pandas as pd

with open('dataset_airbnb_2023-08-17_08-11-16-343.json', 'r', encoding='utf-8') as json_file:
    data = json.load(json_file)
```

```python
#Turning the JSON format into a dataframe
dataset = []
city_values = []

for listing in data:
    amenity_dict = {}
    

    amenity_dict['price'] = listing['pricing_info']["price"]
    amenity_dict['room_type'] = listing['room_type']
    amenity_dict["bedrooms"] = listing["capacity"]["bedrooms"]
    amenity_dict["bathrooms"] = listing["capacity"]["bathrooms"]
    #The data contained 1 rows where the neighbourhood was missing. It was in Bromma, so we can manually add it.
    location = listing["location"]
    if "city" in location:
        city = location["city"]
    else:
        city = "Bromma"
    city_values.append(city)
    
    amenities = listing['amenities']['group']
    for group in amenities:
        for amenity in group['details']:
            amenity_name = amenity['name']
            amenity_available = 1 if amenity['available'] == 'YES' else 0
            amenity_dict[amenity_name] = amenity_available
    
    dataset.append(amenity_dict)

df = pd.DataFrame(dataset)
df = df.fillna(0)
df.insert(1, "neighbourhood", city_values)

df.to_csv('amenities_dataset.csv', index=False)
```

```python
#Checking if it looks right
with pd.option_context('display.max_columns', None):  # more options can be specified also
    display(df)
```
